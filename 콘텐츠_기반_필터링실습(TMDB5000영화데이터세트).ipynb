## 장르 콘텐츠의 유사도 측정

        genres 를 문자열로 변경한 뒤 이를 CountVectorizer로 피처 벡터화한 행렬 데이터 값을 코사인 유사도로 비교

- 문자열로 변환된 genres 칼럼을 Count 기반으로 피처 벡터화 변환
- genres 문자열을 피처 벡터화 행렬로 변환한 데이터 세트를 _**코사인 유사도**_를 통해 비교  
레코드별로 타 레코드와 장르에서 코사인 유사도 값을 가지는 객체를 생성
- 장르 유사도가 높은 영화 중 평점이 높은 순으로 추천


* CountVectorizer \
: 단어들의 카운트(출현 빈도(frequency))로 여러 문서들을 벡터화\
카운트 행렬, 단어 문서 행렬 (Term-Document Matrix, TDM))\
모두 소문자로 변환시키기 때문에 me 와 Me 는 모두 같은 특성이 된다.

```python
# CounterVectorizer : 피처 벡터 행렬


from sklearn.feature_extraction.text import CountVectorizer

# genres 의 리스트를 공배으로 나뉘어진 문자열로 변환

movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))

count_vect = CountVectorizer(min_df = 0 , ngram_range=(1, 2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python
# 코사인 유사도를 통해 비교

from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[0].shape)
print(genre_sim[2].shape)
print(genre_sim[:2])
## 행별 장르 유사도 값을 가지고 있는것 
### 1, 2 => 1행과 2행의 장르 유사도
```

```python
# 유사도가 높은 순으로 다른 레코드를 추춘

genre_sim_soreted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_soreted_ind)

# 각 행별 유사도가 높은 순서대로 추출

## 1행을 보자면 => 가장 높은 유사도는 당연 자기 자신을 제외하고, 3494을 의미한다. 가장 낮은 유사도는 2401
```

### 장르 콘텐츠 필터링을 이용한 영화 추천

```python
## 장르 유사도에 따라 영화 추천


def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    # df 중 title 이 같은 것만 추출
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]



similar_movies = find_sim_movie(movies_df, genre_sim_soreted_ind, 'The Godfather', 10)
similar_movies
```

```python
##평점 높은 순으로 단, 평가자가 1명인 경우를 고려해야 하므로 가중평점을 이요하자 


C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) # 투표 횟수가 많을 수록 가중치 부여 => 상위 60%에게 가중치 부여

print("C :", round(C, 3), 'm : ', round(m, 3))
```

```python
# 가중치를 부여할 상위 percent

percentile = 0.6

C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(percentile)

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )

movies_df['weight_vote'] = movies_df.apply(weighted_vote_average, axis = 1)


```

```python
movies_df[['title', 'vote_average', 'weight_vote', 'vote_count']].sort_values('weight_vote', ascending=False)[:10]
```

```python
## 종합, 장르 유사도 높고 평점 높은 순으로 추천



def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    # df 중 title 이 같은 것만 추출
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)*2]
    similar_indexes = similar_indexes[similar_indexes != title_index] # 자기 자신 제외


    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes].sort_values('weight_vote', ascending=False)[:top_n]




similar_movies = find_sim_movie(movies_df, genre_sim_soreted_ind, 'The Godfather', 10)
similar_movies
```
