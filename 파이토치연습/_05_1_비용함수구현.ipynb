```python
import torch
import torch.nn.functional as F

torch.manual_seed(1)
```

```python
# z = Wx+B

z = torch.FloatTensor([1, 2, 3])
hypothesis = F.softmax(z, dim=0)
hypothesis
```

```python
z = torch.rand(3, 5, requires_grad=True) # 임의의 3x5 행렬 => class가 5개인 경우
z
```

```python
hypothesis = F.softmax(z, dim=1) # dim=1 방향으로 소프트맥스 함수를 적용한다는 의미!
hypothesis
```

```python
hypothesis.sum(dim=1)
```

```python
y = torch.randint(5, (3, )).long() # 0~5 정수중 랜덤하게 3, size 
y
```

```python
y_one_hot = torch.zeros_like(hypothesis)
y_one_hot.scatter_(1, y.unsqueeze(1), 1)

# scatter(dim의 방향, y.unsqueeze(1) 변경할 y 의 위치, 변경할 위치)
# 즉 y의 size(3, ) 에서 y.unsqueeze(1) Dim 1 방향에 1추가 ==> (3, 1)로 만든다.
# y.unsqueeze(1) => 3,1 size로 만들기


# y를 
```
