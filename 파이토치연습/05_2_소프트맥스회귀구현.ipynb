```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

```python
x_train = [[1, 2, 1, 1],
           [2, 1, 3, 2],
           [3, 1, 3, 4],
           [4, 1, 5, 5],
           [1, 7, 5, 5],
           [1, 2, 5, 6],
           [1, 6, 6, 6],
           [1, 7, 7, 7]]
y_train = [2, 2, 2, 1, 1, 1, 0, 0]
x_train = torch.FloatTensor(x_train) # 8x4
y_train = torch.LongTensor(y_train) # 1x3
# LongTensor :  64비트의 부호 있는 정수
```

```python
# 4개의 feature, 8개의 sample, 3개의 class(0, 1, 2)
```

```python
# Label을 One hot encoding 해야함 -> 즉 8개의 라벨을 8,3 size로 One=hot=encoding

y_one_hot = torch.zeros(8, 3)
y_one_hot.scatter_(1, y_train.unsqueeze(1), 1)
y_one_hot
```

```python
# (n_class, 8)size의  z = x@W+ b -> (8, 4)*(4, 3) + (1,3)
# softmax(z)

W = torch.zeros((4, 3), requires_grad=True)
b = torch.zeros((1, 3), requires_grad=True)

optimizer = optim.SGD([W, b], lr=0.1) # 경사하강법~
```

```python
nb_epochs = 1000
for epoch in range(nb_epochs + 1):

    # 가설
    hypothesis = F.softmax(x_train.matmul(W) + b, dim=1) 

    # 비용 함수
    cost = (y_one_hot * -torch.log(hypothesis)).sum(dim=1).mean()

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 100번마다 로그 출력
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

## nn.Moduel 로 구현하기

```python
# 모델의 초기화 ==> 데이터를 보면 4개의 특성 3개의 class 

model = nn.Linear(4, 3)
```

```python
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr=0.1)

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

    # H(x) 계산
    prediction = model(x_train)

    # cost 계산
    cost = F.cross_entropy(prediction, y_train) # 여기에 one_hot_encoding 자동

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 20번마다 로그 출력
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```
